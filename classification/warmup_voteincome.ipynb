````python
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
import pandas as pd
from pydataset import data
data("voteincome", show_doc=True)
````

````python
df = data("voteincome")
````

````python
df = df.drop(columns=["state","year"])
````

````python
df.head()
````

````python
plt.figure(figsize=(12,4))
sns.barplot(x="income",y="vote", hue="female",data=df)
````

````python
X = df[["income","education","age","female"]]
y = df.vote
````

Split the data into train and test datasets. We will be predicting whether or not someone votes based on the the remaining features.

````python
from sklearn.model_selection import train_test_split
````

````python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2)
````

Fit a k-neighbors classifier on the training data. Use 4 for your number of neighbors. How accurate is your model? How does it perform on the test data?

````python
from sklearn.neighbors import KNeighborsClassifier
````

````python
knn = KNeighborsClassifier(n_neighbors=4)
yhat = knn.fit(X_train,y_train)
yhat = yhat.predict(X_train)
````

````python
predictions = pd.DataFrame(data = y_train)
predictions["predicted"] = yhat
predictions.columns = ["actual","predicted"]
````

````python
predictions
````

````python
from sklearn.metrics import classification_report
````

````python
print(classification_report(predictions.actual, predictions.predicted))
````

**What our metrics mean**
- 87% Accurate = Our model can correctly predict 87% of people who will vote with the given features.
- 89% Precision = 89% of the people whom we predicted to vote will actually vote.
- 87% Recall = Out of all our predictions, we correctly predicted 87% of people who will vote.
