## Rename column

## Cast to appropriate data types

### Change to booleans

```python
case = case.withColumn("case_closed",expr('case_closed == "YES"')).withColumn("case_late",expr('case_late == "YES"'))
```

```python
# check if the case_closed and case_late columns were transformed into boolean
case.printSchema()
```

### Change integer-categorical types to string 

```python
case.council_district = case.select(case.council_district.cast("string"))
```

```python
#check to see if we've transformed the data type
case.council_district.dtypes
```

### Change date types

```python
def cast_to_date(df):
    fmt = "M/d/yy H:mm"
    date_cols = ["case_opened_date","case_closed_date","case_due_date"]

    for col in date_cols:
        df = df.withColumn(col, to_timestamp(col,fmt))
    
    return df
```

```python
case = cast_to_date(case)
```

```python
# check if casting has been done
case.printSchema()
```
